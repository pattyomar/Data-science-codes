```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\My Python codes\SouthAfricaCrimeStats_v2.csv')
df
```

```python
#to check for null values, no null values found
df.info()
```

```python
print(list(df['Province'].unique()))
print(df['Station'].unique())
print(df['Category'].unique())

```

```python
df_province = df.groupby(['Province']).sum()
df_province
```

```python
df_province.transpose().describe().round(0)
```

```python
df_province.plot(kind='bar', rot=80)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=1)
plt.title('Crime per province')
#Graph to show provinces with highest crime rates
```

```python
df_province.plot(kind='bar', figsize = (8, 50), subplots = True, rot=80)
plt.legend()
plt.title('Crime per province')
```

```python
df_crimes = df.groupby(['Category']).sum()
df_crimes
```

```python
df_crimes.plot(kind='bar', figsize = (15, 6), rot=90)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
plt.title('Common crimes')
#Graph to show most common crimes in SA
```

```python
df_crimes.plot(kind='bar', figsize = (10, 50), subplots = True, rot=90)
plt.legend()
plt.title('Common crimes')
```

```python
western_cape = df[df["Province"]=="Western Cape"]
gauteng = df[df["Province"]=="Gauteng"]
free_state = df[df["Province"]=="Free State"]
n_west = df[df["Province"]=="North West"]
kwazulu = df[df["Province"]=="Kwazulu/Natal"]
mpumalanga = df[df["Province"]=="Mpumalanga"]
eastern_cape = df[df["Province"]=="Eastern Cape"]
limpopo = df[df["Province"]=="Limpopo"]
northern_cape = df[df["Province"]=="Northern Cape"]

```

```python
western_cape = df[df["Province"]=="Western Cape"].sum
western_cape
```

```python
common_crime = western_cape.groupby(['Category']).sum()
common_crime
```

```python
common_crime.plot(kind='bar', figsize = (15, 6), rot=90)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
plt.title('Common crimes in Western cape')
```

```python
cc_pct = common_crime.pct_change(axis = 'columns' ).fillna(0).round(3)
cc_pct
#to claculate percentage change between columns
#remove the'axis' parameter to calculate % change between rows
```

```python
cc_pct1 = cc_pct.transpose()
cc_pct1
```

```python
cc_pct1.plot(kind='line', figsize = (25, 6), rot=90)
plt.legend(ncol = 3)
plt.title('change over the years')
#Graph to show percentage change in different crimes over the years in Western Cape Province
```

```python
cc_pct1.plot(kind='line', figsize = (10, 50), subplots = True, rot=90)
plt.legend()
plt.title('change over the years')
```
